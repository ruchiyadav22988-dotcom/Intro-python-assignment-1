2. What is an interpreter in Python?

   -An interpreter in Python (or any interpreted language) is a program that reads and executes code line by line at runtime, without the need to compile the entire code beforehand.

3. What are pre-defined keywords in Python?

   -In Python, keywords (also called reserved words) are words that have special meaning and are reserved for specific purposes in the language. These cannot be used as variable names, function names, or identifiers.
4. Can keywords be used as variable names?

   -Keywords are reserved words that have a special meaning in the language, such as for, while, if, class, def, True, False, etc. If you try to use them as variable names, Python will throw a SyntaxError.
5. What is mutability in Python?

   -Mutability in Python refers to whether or not an object’s value can be changed after it is created. Mutable objects → Can be changed (modified) in place.Immutable objects → Cannot be changed once created; any “change” creates a new object.
6. Why are lists mutable, but tuples are immutable?

   -A list is designed to be a flexible data structure where elements can be added, removed, or modified.Internally, Python stores a list as an array of references (pointers) to objects.

     -== → Equality operator Checks whether the values of two objects are the same.It doesn’t care if they are stored in different memory locations.

10. What is the difference between implicit and explicit type casting?

     -Implicit casting → Python handles it safely.Explicit casting → You handle it manually with conversion functions.

11. What is the purpose of conditional statements in Python?

     -The main purpose of conditional statements is to let a program make decisions and execute code selectively based on conditions.

12. How does the elif statement work?

     -It lets you check multiple conditions one after another.python executes the first condition that is True and then skips the rest.If none of the conditions are True, the else block runs.

13. What is the difference between for and while loops?

     -For Loop Used when you know exactly how many times you want to repeat something.Iterates over a sequence (like list, tuple, string, range, etc.).While Loop Used when you want to repeat until a condition is no longer True.The number of iterations is not fixed beforehand.
